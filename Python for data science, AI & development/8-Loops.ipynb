```python
# Example of range
print("range(3):", range(3))

# Example of for loop with release_years
release_years = [1982, 1980, 1973]
num_years = len(release_years)

for index in range(num_years):
    print(release_years[index])

# Example of for loop with range
for index in range(0, 8):
    print(index)

# Example of for loop, loop through list
for current_year in release_years:
    print(current_year)

# Use for loop to change the elements in list
color_list = ['red', 'yellow', 'green', 'purple', 'blue']

for index in range(0, 5):
    print("Before square", index, "is", color_list[index])
    color_list[index] = 'white'
    print("After square", index, "is", color_list[index])

# Loop through the list and iterate on both index and element value
for index, color in enumerate(color_list):
    print(index, color)

# While Loop Example
release_years = [1982, 1980, 1973, 2000]
index = 0
current_year = release_years[0]

while current_year != 1973:
    print(current_year)
    index = index + 1
    current_year = release_years[index]

print("It took", index, "repetitions to get out of loop.")

# Quiz on Loops
# Print elements between -5 and 5
for index in range(-4, 5):
    print(index)

# Print elements of music_genres
music_genres = ['rock', 'R&B', 'Soundtrack', 'R&B', 'soul', 'pop']
for genre in music_genres:
    print(genre)

# Print elements of color_list
color_list = ['red', 'yellow', 'green', 'purple', 'blue']
for color in color_list:
    print(color)

# While loop to display album ratings
album_ratings = [10, 9.5, 10, 8, 7.5, 5, 10, 10]
index = 0
current_rating = album_ratings[0]

while index < len(album_ratings) and current_rating >= 6:
    print(current_rating)
    index = index + 1
    current_rating = album_ratings[index]

# While loop to copy 'orange' to new_color_list
fruit_list = ['orange', 'orange', 'purple', 'blue ', 'orange']
new_color_list = []
index = 0

while index < len(fruit_list) and fruit_list[index] == 'orange':
    new_color_list.append(fruit_list[index])
    index = index + 1

print(new_color_list)
```
